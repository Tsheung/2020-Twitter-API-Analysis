I created a class named Streamer to stream the first 1000 tweets without Geotags. I wanted to just examine the number of tweets that mentioned a certain candidate. The first function filters the tweet as English and prints out the current number of tweets recieved. I wanted to collect only 1000 tweets first. The second function allows the program to disconnect if there are errors while streaming. The class appends the tweets into a list called my_tweet.  

```python
my_tweets = []
```

```python
from twython import TwythonStreamer

class Streamer(TwythonStreamer):
    
    def on_success(self, data):
        if data['lang'] == 'en':
            my_tweets.append(data)
            print('Received tweet #' + str(len(my_tweets)))
        
        if len(my_tweets) >= 1000:
            self.disconnect()
    
    def on_error(self, status_code, data):
        print(status_code, data)
        self.disconnect()
```

Below, I analyzed the tweets inside the my_tweets json file. I wanted to initially to examine whether or not my function worked and whether or not the file actually contained the tweets. The tweets['text'] examines the tags the user tagged the candidates and what they tweeted about. The tweets['user']['location'] is the user's location in their profile.  

```python
import json 
from pprint import pprint
with open('my_tweets.json', 'r') as my_file:
    data_json = my_file.read()
    this_data = json.loads(data_json)
    
for tweets in this_data:
    print(tweets['text'])
    print(tweets['user']['location'])
```

I created a list for each candidate to contain each of the tweet's locations. The code below uses a if loop if the tweet['text'] contains either candidate or more and appends that tweets' location to the list. After the loop, I decided to examine the counts of each candidate. 

```python
#Analyze the location to city, state
#isolate by candidate then save the location/place
s = [] #sanders
w = [] #warren
b = [] #biden


for tweets in this_data:
    if '@BernieSanders' in tweets['text'] and '@ewarren' in tweets['text'] and '@JoeBiden' in tweets['text']:
        s.append(tweets['user']['location'])
        w.append(tweets['user']['location'])
        b.append(tweets['user']['location'])
        
    elif '@BernieSanders' in tweets['text'] and '@ewarren' in tweets['text']:
        s.append(tweets['user']['location'])
        w.append(tweets['user']['location'])
        
    elif '@JoeBiden' in tweets['text'] and '@ewarren' in tweets['text']:
        b.append(tweets['user']['location'])
        w.append(tweets['user']['location'])
        
    elif '@BernieSanders' in tweets['text'] and '@JoeBiden' in tweets['text']:
        s.append(tweets['user']['location'])
        b.append(tweets['user']['location'])
        
    elif '@BernieSanders' in tweets['text']:
        s.append(tweets['user']['location'])
        
    elif '@ewarren' in tweets['text']:
        w.append(tweets['user']['location'])
        
    elif '@JoeBiden' in tweets['text']:
        b.append(tweets['user']['location'])

print(len(s))
print(len(w))
print(len(b))
```

It seems that Joe Biden is the candidate that is most tweeted about. After is Bernie Sander and then Elizabeth Warren.

Below, I decided to examine the most common locations for each candidate. I used a Counter function and printed out the results. As I can see, some of the locations are not defined by state and some are fake locations. 

```python
from collections import Counter

#Analyze by state #Sanders
word_count = Counter()
for state in s:
    if state != None:
        word_count[state] += 1
print(word_count.most_common())
```

```python
#Warren 
word_count = Counter()
for state in w:
    if state != None:
        word_count[state] += 1
print(word_count.most_common())
```

```python
#Biden
word_count = Counter()
for state in b:
    if state != None:
        word_count[state] += 1
print(word_count.most_common())
```

It is noticable that the user's location that mention Joe Biden is quite an extensive list.
